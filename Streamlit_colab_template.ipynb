```
%%writefile app.py
import streamlit as st
import investpy
import pandas as pd
import plotly.express as px
import datetime
from matplotlib import pyplot as plt
from fbprophet import Prophet
from fbprophet.plot import plot_plotly

@st.cache
def start():
    today=datetime.date.today().strftime('%d/%m/%Y')
    commodity_ = investpy.commodities.get_commodities_list(group=None)

    asset = st.sidebar.selectbox('Click below to select a new asset',
                                    commodity_)

    commo = investpy.get_commodity_historical_data(commodity="GOLD", 
                                                from_date='01/01/1980', 
                                                to_date=today,
                                                as_json=False,
                                                order='ascending')
    data = commo.copy()
    data.index.name = None

    return data

data = start()
st.dataframe(data)

section = st.sidebar.slider('Number of observation', 
                            min_value=30,
                            max_value=min([2000, 4000]),
                            value=500,  
                            step=10)
```
